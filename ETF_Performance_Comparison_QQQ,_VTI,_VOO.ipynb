```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, NumeralTickFormatter
from bokeh.transform import factor_cmap

etfs = ["QQQ", "VTI", "VOO"]

volality_fi = figure(
    title="Average Daily Trading Volume (3 ETFs)",
    x_range=etfs,
    x_axis_label="ETF",
    y_axis_label="Average Volume",
    height=300,
    width=800
)

source = ColumnDataSource(data={
    "etf": etfs,
    "volume": [qqq_mean_vol, vti_mean_vol, voo_mean_vol]
})

volality_fi.vbar(
    x="etf",
    top="volume",
    source=source,
    width=0.6,
    fill_color = factor_cmap("etf",
        palette=colors,
        factors=etfs)
)

volality_fi.yaxis.formatter = NumeralTickFormatter(format="0.0a")

show(volality_fi)
```

```python
import pandas as pd

qqq_df['Date'] = pd.to_datetime(qqq_df['Date'])
vti_df['Date'] = pd.to_datetime(vti_df['Date'])
voo_df['Date'] = pd.to_datetime(voo_df['Date'])
```

```python
qqq_df['Peak'] = qqq_df['Close'].cummax()
vti_df['Peak'] = vti_df['Close'].cummax()
voo_df['Peak'] = voo_df['Close'].cummax()
```

```python
for df in [qqq_df, vti_df, voo_df]:
    df['Month'] = df['Date'].dt.strftime('%b')
    df['Year'] = df['Date'].dt.year
```

```python
for df in [qqq_df, vti_df, voo_df]:
    df['Drawdown'] = (df['Close'] - df['Peak']) / df['Peak'] * 100
```

```python
import pandas as pd
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure, show
from bokeh.transform import factor_cmap

months_order = ["Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

qqq_monthly = qqq_monthly
vti_monthly = vti_monthly
voo_monthly = voo_monthly

factors = [(month, etf) for month in months_order for etf in ["QQQ", "VTI", "VOO"]]

drawdowns = []
for q, v, vo in zip(qqq_monthly, vti_monthly, voo_monthly):
    drawdowns.extend([q, v, vo])

source = ColumnDataSource(data=dict(
    x=factors,
    drawdown=drawdowns
))

colors = ["blue", "green", "red"]
```

```python
drawdowns_fig = figure(
    x_range=FactorRange(*factors),
    height=400, width=900,
    title="Monthly Worst Drawdowns (%)",
    y_axis_label="Drawdown %"
)

drawdowns_fig.vbar(
    x='x',
    top='drawdown',
    width=0.8,
    source=source,
    line_color="white",
    fill_color=factor_cmap('x', palette=colors, factors=["QQQ", "VTI", "VOO"], start=1, end=2)
)

drawdowns_fig.xaxis.major_label_orientation = 1
drawdowns_fig.xgrid.grid_line_color = None
drawdowns_fig.y_range.end = 0

show(drawdowns_fig)
```
